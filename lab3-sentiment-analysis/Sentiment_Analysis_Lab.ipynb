# Zadania Laboratoryjne: Analiza Sentymentu i Przetwarzanie Języka Naturalnego na Tweetach

Poniższe zadania mają na celu przećwiczenie umiejętności zbierania danych z Twittera, analizy sentymentu, przetwarzania językowego oraz rozpoznawania encji nazwanych (NER) przy użyciu bibliotek takich jak **Tweepy**, **spaCy** oraz **Stanza**.

---

## Zadanie 1: Pobieranie Tweetów z Tweepy

**Cel:**  
Pobierz dane z Twittera z wybranego konta (np. Kanye West, Elon Musk, Donald Trump) przy użyciu biblioteki Tweepy.

**Kroki:**
- Zarejestruj aplikację na [Twitter Developer Platform](https://developer.twitter.com/) i uzyskaj klucze API.
- Wybierz profil użytkownika, który regularnie publikuje tweety.
- Pobierz określoną liczbę tweetów (np. 100) i zapisz je do pliku (CSV lub JSON) w celu dalszej analizy.
- Zadbaj o przestrzeganie limitów API i polityki Twittera.

---

## Zadanie 2: Analiza Sentymentu Tweetów przy użyciu spaCy

**Cel:**  
Wykorzystaj bibliotekę spaCy do przeprowadzenia analizy sentymentu pobranych tweetów.

**Kroki:**
- Wczytaj dane z pliku uzyskane w Zadaniu 1.
- Zaimplementuj analizę sentymentu tweetów. Możesz:
  - Zastosować gotowy model lub rozszerzyć pipeline spaCy o dodatkowe reguły,
  - Alternatywnie, połączyć spaCy z innymi narzędziami (np. TextBlob) dla oceny sentymentu.
- Przypisz każdemu tweetowi etykietę: **pozytywny**, **neutralny** lub **negatywny**.
- Zwizualizuj wyniki, np.:
  - Wykres kołowy przedstawiający procentowy rozkład sentymentu.
  - Wykres słupkowy z liczbą tweetów dla każdej kategorii.

---

## Zadanie 3: Analiza Językowa i Sentymentu przy użyciu Stanza

**Cel:**  
Przeprowadź analizę językową tweetów oraz oblicz sentyment przy użyciu biblioteki Stanza.

**Kroki:**
- Wczytaj te same dane tweetów.
- Użyj Stanza do:
  - Lematizacji tweetów,
  - Analizy składniowej (wyznaczenie rdzenia zdania, części mowy itp.).
- Oblicz sentyment tweetów korzystając z dostępnych narzędzi lub własnych reguł.
- Porównaj wyniki sentymentu uzyskane przy użyciu Stanza z wynikami z Zadania 2 (spaCy).
- Zwizualizuj porównanie wyników (np. wykresy porównawcze).

---

## Zadanie 4: Rozpoznawanie Encji Nazwanych (NER)

**Cel:**  
Wykonaj rozpoznawanie encji nazwanych (NER) na tweetach, korzystając z dwóch metod: Stanza oraz własnej implementacji (np. z użyciem spaCy).

**Kroki:**
- Przeprowadź NER na tweetach przy użyciu Stanza.
- Zaimplementuj własne rozwiązanie NER wykorzystujące spaCy (możesz użyć pretrenowanego modelu).
- Zwizualizuj wyniki obu metod, np.:
  - Oryginalne tweety z wyróżnionymi encjami (osoby, organizacje, lokalizacje).
  - Tabela z zestawieniem wykrytych encji.
- Porównaj efektywność obu podejść i omów różnice w wynikach.

---

## Zadanie 5: Klasyfikacja Sentymentu na Innym Datasetcie oraz Dodatkowy Proces NER

**Cel:**  
Zastosuj dowolną metodę klasyfikacji sentymentu na wybranym zbiorze danych oraz przeprowadź NER przy użyciu biblioteki (spaCy lub Stanza). Następnie porównaj wyniki własnego klasyfikatora z wynikami analizy sentymentu uzyskanymi przy użyciu Stanza.

**Kroki:**
- Wybierz dowolny dataset do analizy sentymentu (np. recenzje filmowe, opinie o produktach).
- Zaimplementuj klasyfikator sentymentu wykorzystując wybraną metodę (np. SVM, drzewa decyzyjne lub sieci neuronowe).
- Przeprowadź proces NER na tym datasetcie, korzystając z biblioteki spaCy lub Stanza.
- Porównaj:
  - Wyniki klasyfikacji sentymentu własnego klasyfikatora,
  - Wyniki analizy sentymentu uzyskane przez Stanza.
- Zwizualizuj wyniki (np. wykresy porównawcze, tabele) i opisz uzyskane różnice.
